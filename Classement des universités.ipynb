```python
annees_cwur = df_cwur['year'].value_counts()
print ("Pour le ranking du CWUR, on a :")
print("- Le nombre d'années pour lesquelles un classement a été fait :", annees_cwur.shape[0])
print("- Les années pour lesquelles un classement a été fait :", sorted(annees_cwur.index.tolist()))
print("- Le nombre moyen d'universités incluses dans le classement :", round(annees_cwur.mean(), 0))
print("- Le détail du nombre d'universités incluses dans le classement par année :\n", annees_cwur.sort_index())
```

```python
annees_times = df_times['year'].value_counts()
print ("Pour le ranking du Times, on a :")
print("- Le nombre d'années pour lesquelles un classement a été fait :", annees_times.shape[0])
print("- Les années pour lesquelles un classement a été fait :", sorted(annees_times.index.tolist()))
print("- Le nombre moyen d'universités incluses dans le classement :", round(annees_times.mean(),0))
print("- Le détail du nombre d'universités incluses dans le classement par année :\n", annees_times.sort_index())
```

```python
annees_shanghai = df_shanghai['year'].value_counts()
print ("Pour le Shanghai ranking, on a :")
print("- Le nombre d'années pour lesquelles un classement a été fait :", annees_shanghai.shape[0])
print("- Les années pour lesquelles un classement a été fait :", sorted(annees_shanghai.index.tolist()))
print("- Le nombre moyen d'universités incluses dans le classement :", round(annees_shanghai.mean(),0))
print("- Le détail du nombre d'universités incluses dans le classement par année :\n", annees_shanghai.sort_index())
```

```python
index_strasbourg_cwur = df_cwur.index[df_cwur['institution'].str.contains('Strasbourg')].tolist()
print("Voici les rankings obtenus par l'Université de Strasbourg dans le classement du CWUR :")
ranking_strasbourg_cwur = []
for i in index_strasbourg_cwur:
    ranking_strasbourg_cwur.append(df_cwur.iloc[i]['world_rank'])
print(ranking_strasbourg_cwur)
```

```python
index_strasbourg_times = df_times.index[df_times['university_name'].str.contains('Strasbourg')].tolist()
print("Voici les rankings obtenus par l'Université de Strasbourg dans le classement du Times :")
ranking_strasbourg_times = []
for i in index_strasbourg_times:
    ranking_strasbourg_times.append(df_times.iloc[i]['world_rank'])
ranking_strasbourg_times.sort()
print(ranking_strasbourg_times)
```

```python
index_strasbourg_shanghai = df_shanghai.index[df_shanghai['university_name'].str.contains('Strasbourg', na=False)].tolist()
print("Voici les rankings obtenus par l'Université de Strasbourg dans le classement de Shanghai :")
ranking_strasbourg_shanghai = []
for i in index_strasbourg_shanghai:
    ranking_strasbourg_shanghai.append(df_shanghai.iloc[i]['world_rank'])
ranking_strasbourg_shanghai.sort(key=len)
print(ranking_strasbourg_shanghai)
```

```python
ranking_strasbourg_times_1 = ['213' if '201-225' in s else '325.5' for s in ranking_strasbourg_times]
ranking_strasbourg_shanghai_1 = ['126' if '101-151' in s else s for s in ranking_strasbourg_shanghai]
ranking_strasbourg_shanghai_2 = ['126' if '102-150' in s else s for s in ranking_strasbourg_shanghai_1]
ranking_strasbourg_shanghai_3 = ['125.5' if '101-150' in s else s for s in ranking_strasbourg_shanghai_2]
joined_rankings_strasbourg_str = ranking_strasbourg_cwur + ranking_strasbourg_times_1 + ranking_strasbourg_shanghai_3
joined_rankings_strasbourg_int = list(map(float, joined_rankings_strasbourg_str))
av_rank_strasbourg = sum(joined_rankings_strasbourg_int) / len(joined_rankings_strasbourg_int)
print("En prenant uniquement en compte les classements où elle apparaît, l'Université de Strasbourg a un rang moyen de", int(round(av_rank_strasbourg,0)))
```

```python
df_times_2016 = df_times.loc[(df_times['year'] == 2016)]
print("Voici les 20 meilleures universités en 2016 selon le classement du Times :")
df_times_2016[['world_rank', 'university_name']].head(20).style.hide_index()
```

```python
df_times_2016_usa = df_times_2016['country'].value_counts() / df_times_2016['country'].value_counts().sum()
df_times_2016_usa = df_times_2016_usa.to_frame(name='repartition')
df_times_2016_usa.loc[:, 'repartition'] = df_times_2016_usa['repartition'].map('{:.2%}'.format)
print("Voici la répartition par pays des universités présentes dans le classement 2016 du Times :")
df_times_2016_usa
```

```python
print("En 2016, il y avait", df_times_2016_usa._get_value('United States of America', 'repartition'), "d'universités américaines dans le classement du Times.") 
```

```python
df_shanghai_2015 = df_shanghai[['university_name', 'year']].dropna(axis=0, how='any')
df_shanghai_2015 = df_shanghai_2015.query("year == 2015").drop('year', axis=1)

df_table_school_country = pd.read_csv("school_and_country_table.csv").rename(columns={'school_name': 'university_name'})

merge_shanghai_2015 = pd.merge(df_shanghai_2015, df_table_school_country, how="left")
print("Voici combien chaque pays possède d'universités dans le classement de Shanghai en 2015 :")
merge_shanghai_2015['country'].value_counts().sort_values(ascending=False)
```

```python
df_cwur_france = df_cwur.loc[df_cwur['country'] == 'France']
schools_cwur_france = np.sort(df_cwur_france.institution.unique())
print("Les universités françaises présentes dans le classement CWUR sont les suivantes :\n\n", schools_cwur_france)
```

```python
df_times_2011 = df_times.loc[(df_times['year'] == 2011)]
nb_by_country_times_2011 = df_times_2011['country'].value_counts().to_frame(name='nb_schools')
nb_by_country_times_2011 = nb_by_country_times_2011.reset_index().rename(columns={'index': 'country'})

df_edu_expenditure_2011 = df_edu_expenditure[['country', 'institute_type', 'direct_expenditure_type', '2011']]
df_edu_expenditure_2011 = df_edu_expenditure_2011[~df_edu_expenditure_2011.institute_type.str.contains("All Institutions|Elementary and Secondary Institutions")]
df_edu_expenditure_2011 = df_edu_expenditure_2011[df_edu_expenditure_2011.direct_expenditure_type.str.contains("Total")]

list_missing_countries=[]
for country in nb_by_country_times_2011.country.tolist():
    if country not in df_edu_expenditure_2011.country.tolist():
        list_missing_countries.append(country)
print("Voici les pays du df_times_2011 qui sont manquants ou qui ne sont pas écrits de la même manière dans df_edu_expenditure :\n", list_missing_countries)
df_edu_expenditure_2011 = df_edu_expenditure_2011.replace({'country':{'United States' : 'United States of America', 'Ireland' : 'Republic of Ireland', 'Korea, Republic of' : 'South Korea'}})
print("Nous allons homogénéiser les noms de pays et supprimer les lignes pour lesquelles nous n'avons pas de valeur.")

merge_times_2011_expenditures = pd.merge(nb_by_country_times_2011, df_edu_expenditure_2011, how="left").dropna().rename(columns={'2011': 'expenditures_2011'})
merge_times_2011_expenditures
```

```python
ratio = merge_times_2011_expenditures['expenditures_2011']/merge_times_2011_expenditures['nb_schools']
ratio = ratio.round(decimals=2)
merge_times_2011_expenditures['ratio'] = ratio
merge_times_2011_expenditures.sort_values(by=['ratio'])
```
